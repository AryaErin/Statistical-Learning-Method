```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
fig_path = './lena.jpg'
img = cv2.imread(fig_path)
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.figure(figsize = (10,10))
plt.title('original image')
plt.imshow(img, cmap = 'gray')
print(img.shape)
```

```python
u, s, vh = np.linalg.svd(img)
```

```python
# 选取前k个奇异值对图像进行重构
def imshow(u, s, vh, k):
    new_image = u[:,:k] @ np.diag(s[:k]) @ vh[:k,:]
    plt.figure(figsize = (10, 10))
    plt.title('using previous %d singular values to reconstruct image.' % (k))
    plt.imshow(new_image, cmap = 'gray')
```

```python
imshow(u, s, vh, 200)
imshow(u, s, vh, 300)
imshow(u, s, vh, 400)
```
